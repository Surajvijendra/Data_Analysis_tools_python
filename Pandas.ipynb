#### Join and Append in pandas
* Use Suffixes: specify suffixes to differentiate the overlapping columns in the two DataFrames if columns having same name.
* if both table having different name then it will exicute without any suffix.

```python
t1.join(t2, how='right', lsuffix='_left', rsuffix='_right')
```

```python
t1.join(t2, how='left', lsuffix='_left', rsuffix='_right')
```

### Left Join (how='left')
* A left join includes all rows from the left DataFrame (t1 in this case) and only matching rows from the right DataFrame (t2).
* if a row from the left DataFrame has no match in the right DataFrame, it will still be included, with NaN values in the columns from the right DataFrame.
* Rows in the right DataFrame without a match in the left DataFrame are ignored.

### Right Join (how='right')
* A right join includes all rows from the right DataFrame (t2), and only matching rows from the left DataFrame (t1).

* If a row in the right DataFrame has no match in the left DataFrame, it will still be included, with NaN values in the columns from the left DataFrame.
* Rows in the left DataFrame without a match in the right DataFrame are ignored.

### Inner Join (how='inner')
* An inner join only includes rows that have matching indices (or columns, if using pd.merge()) in both DataFrames.

* Rows with no match in either DataFrame are excluded.
* This join type results in the smallest number of rows among the join types because it only keeps rows that are present in both DataFrames.

###  Outer Join (how='outer')
* An outer join includes all rows from both DataFrames, filling in NaN where there are missing matches.

* If a row in one DataFrame has no corresponding match in the other DataFrame, it will still be included, with NaN values for the columns of the missing DataFrame.
* This join type results in the largest number of rows because it includes all rows from both DataFrames.

#### Key Differences between merge and join

* .join() is mainly for joining DataFrames based on their indices.

* pd.merge() is more flexible, allowing you to join DataFrames based on columns (or on indices if specified).

#### Pivot table

```python
import pandas as pd

# Example DataFrame
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02'],
    'Category': ['A', 'A', 'B', 'B'],
    'Sales': [100, 150, 200, 250],
    'Quantity': [10, 15, 20, 25]
}

df = pd.DataFrame(data)

# Creating a pivot table
pivot_df = df.pivot_table(
    values='Sales',         # What to aggregate
    index='Date',           # Row labels
    columns='Category',     # Column labels
    aggfunc='sum',          # Aggregation function
    fill_value=0            # Filling NaN with 0
)

print(pivot_df)
```
