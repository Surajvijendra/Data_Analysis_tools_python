### Label-based locator and  Integer location-based locator

<pre>Key Differences:
loc: Works with labels. Example: df.loc[0:2, 'column_name']
iloc: Works with numerical positions. Example: df.iloc[0:2, 1:3] </pre>

* in order to get the rows between 2 to 9 and the columns between 2nd to 5th

* in order to get or skip few rows or columns while indexing
* alternative rows we will be getting and also  columns

### Group by and Aggregating

<pre>Common Aggregations in pandas
    
count(): Counts non-NA entries in each group.
sum(): Sums up the values in each group.
mean(): Calculates the average of each group.
median(): Calculates the median.
min()/max(): Finds the minimum/maximum.
std()/var(): Standard deviation and variance.
prod(): Calculates the product of values.
first()/last(): Returns the first/last value in each group.
nunique(): Counts distinct values in each group</pre>

```python
raw_data.groupby('smoker').count()
```

### Strips
* The strip() function in pandas is used to remove leading and trailing characters (including whitespace) from string elements in a column. This is particularly helpful for cleaning data, such as removing unwanted spaces or special characters from text data.
* It typically apply strip() to a pandas Series or DataFrame column containing string data using the '.str' accessor.

#### 1. Basic Stripping
* Removes leading and trailing whitespace (default behavior).
* df['column'] = df['column'].str.strip()

```python
raw_data['region'].str.strip()
```

#### 2. Stripping Specific Characters
* Removes specified characters (e.g., $, #, or *) from both ends.


```python
raw_data['region'].str.strip('!@#$%')
```

#### 3. Removing Only Leading Characters (lstrip)
* Removes characters only from the beginning of the string.

```python
raw_data['region'].str.lstrip('!@#$%')
```

#### 4. Removing Only Trailing Characters (rstrip)
* Removes characters only from the end of the string.

```python
raw_data['region'].str.rstrip('!@#$%')
```

#### 5. Handling Non-String Columns
* Convert the column to string type before applying strip().

```python
raw_data['bmi'].astype(str).str.strip()
```
