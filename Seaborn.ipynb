#### Seaborn, a powerful Python visualization library built on top of Matplotlib, offers various types of plots for statistical data visualization. Some commonly used types of plots in Seaborn include:

<pre>1. Relational Plots (For relationships between variables)
scatterplot(): Displays a scatter plot.
lineplot(): Displays a line plot.
relplot(): A figure-level function to create scatter or line plots.
    
2. Categorical Plots (For categorical data)
stripplot(): Draws a scatter plot of categorical data.
swarmplot(): Similar to stripplot() but adjusts points to avoid overlap.
boxplot(): Shows a box-and-whisker plot.
violinplot(): Combines a box plot and KDE plot.
boxenplot(): Similar to boxplot() but better for large datasets.
pointplot(): Shows mean values with confidence intervals.
barplot(): Displays a bar chart with aggregated values.
countplot(): Shows the count of occurrences in each category.
    
3. Distribution Plots (For univariate/multivariate distributions)
histplot(): Creates a histogram.
kdeplot(): Displays a Kernel Density Estimate plot.
ecdfplot(): Plots the Empirical Cumulative Distribution Function (ECDF).
rugplot(): Draws marks along an axis to show data distribution.
    
4. Regression Plots (For trend analysis)
lmplot(): Displays a linear regression plot.
regplot(): Similar to lmplot() but works on a single axis.
residplot(): Shows residuals of a regression model.
    
5. Matrix Plots (For correlation and clustering)
heatmap(): Displays a heatmap of a matrix (e.g., correlation matrix).
clustermap(): Performs hierarchical clustering and visualizes it as a heatmap.
    
6. Multivariate Plots (For multiple variables)
pairplot(): Plots pairwise relationships between multiple numerical variables.
jointplot(): Shows bivariate distribution with scatter/hex/kde plots.
    
7. Timeseries Plots (For temporal data)
lineplot(): Used for visualizing trends over time. </pre>

#### Violin Plot
* It's a combination of a box plot and a kernel density plot. It is used to visualize the distribution of a dataset and provides insights into the data's spread, skewness, and probability density.
* It uses Kernel Density Estimation (KDE) to create a smooth curve that represents how frequently values occur.
* Unlike a box plot, which only shows quartiles and outliers, a violin plot visualizes the entire distribution.
* Shows where most data points are concentrated (wider areas = higher density).
* Can capture multi-modal distributions (box plots cannot).
* A multimodal distribution is a probability distribution that has multiple peaks (or modes). This means the data is concentrated around more than one value, rather than following a single central tendency like a normal (unimodal) distribution.


parameter
* bw --> band width
* cut -->
* scale --> Adjusts the width of the violins.

#### Basic Plot

```python
sns.violinplot(x='survived',y='age',data=raw_data)
plt.show()
```

```python
sns.violinplot(x='survived',y='age',data=raw_data,hue='sex')
plt.show()
```

##### Using split
* If hue is specified, setting split=True will show both categories on the same violin instead of separate violins.

```python
sns.violinplot(x='survived',y='age',data=raw_data,hue='sex',split=True)
plt.show()
```

#### inner
Controls the internal representation inside the violin.
* box --> Shows a mini box plot inside the violin.
* quartile --> Marks the quartiles.
* points --> Shows individual data points.
* stick --> Shows a stick representation of the KDE.

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x="pclass", y="age", data=raw_data, scale="count", inner="quartile")
plt.title("Age Distribution Across Passenger Classes")
plt.show()
```

#### Scale 
* area (default) --> All violins have the same area.
* count --> Violin width is proportional to the number of observations.
* width --> All violins have the same maximum width.

#### Cut
Determines how far the KDE extends beyond the minimum and maximum values.

```python
sns.violinplot(x="pclass", y="age", data=raw_data, scale="width")
plt.show()
```

```python
sns.violinplot(x="pclass", y="age", data=raw_data, cut = 0)
plt.show()
```

#### Kde Plot
* A KDE (Kernel Density Estimation) plot is a smooth approximation of a histogram that represents the probability density function (PDF) of a continuous variable. It helps visualize the distribution of data in a continuous manner by estimating the probability density function.
* Understanding the Distribution: Helps visualize the shape of data distribution.
* Smoothing Histograms: Provides a smooth alternative to histograms.
* Comparing Multiple Distributions: Useful for comparing different datasets.

#### Kde Plot Parameters
* bw_adjust --> Controls the smoothness of the KDE curve.
* A lower value (<1) makes the curve more sensitive (less smooth).
* A higher value (>1) makes it smoother.

```python
sns.kdeplot(x='age',data=raw_data,hue='sex',bw_adjust=1)
plt.show()
```

#### DisPlot
* displot in Seaborn (short for "distribution plot") is used to visualize the distribution of a dataset. It is a figure-level function that can generate different types of distribution plots, including histograms, kernel density estimates (KDE), and empirical cumulative distribution functions (ECDF).
* data --> The dataset to be plotted.
* bins --> Number of bins in the histogram.
* kde --> If True, overlays a Kernel Density Estimate plot.
* rug --> If True, adds a rug plot at the bottom.
* kind --> Can be 'hist' (default), 'kde', or 'ecdf' to specify the type of plot.
* col --> will get the plot on different column based on category 

```python
sns.displot(x='age',data=raw_data,kde=True,row='sex',hue='sex',col='pclass')
plt.show()
```

#### Pairplot
* A pair plot in Seaborn is used to visualize pairwise relationships between numerical features in a dataset. It is particularly useful for exploratory data analysis (EDA) because it helps identify trends, correlations, and patterns between variables.
* Visualizes Relationships: Shows scatter plots between pairs of numerical variables.
* Identifies Correlations: Helps in spotting relationships (positive, negative, or no correlation).
* Detects Outliers: Outliers can be easily spotted in scatter plots.
* Class Separation: If a categorical variable is used (hue parameter), it helps see how different categories are distributed.

<pre>sns.pairplot(

)

#### Relplot
* A relplot (Relational Plot) in Seaborn is used to visualize relationships between two numerical variables. It is a high-level function that can create both scatter plots and line plots using the kind parameter.
* When you want to easily switch between scatter and line plots.

```python
sns.relplot(data=flights,y='year',x='passengers',hue='month',kind='scatter')
plt.show()
```

#### Regplot
* Regplot is used to plot a scatterplot with a regression line in Seaborn. It’s useful for visualizing linear relationships between two continuous variables.
* scatter_kws: Dictionary of keyword arguments for scatter points (e.g., {"color": "red"}).
* line_kws: Dictionary of keyword arguments for the regression line (e.g., {"color": "blue", "lw": 2}).
* fit_reg: Set to False to remove the regression line.

```python
sns.regplot(y='fare',x='sex',data=raw_data)
plt.show()
```
