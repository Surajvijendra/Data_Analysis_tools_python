####  Grouped Bar Plot
* Used to compare multiple groups side by side.
* Two sets of bars are plotted next to each other for comparison.
* The x-axis positions are adjusted using x - width/2 and x + width/2.


```python
import numpy as np

categories = ['A', 'B', 'C', 'D']
group1 = [10, 15, 7, 12]
group2 = [8, 13, 6, 9]

x = np.arange(len(categories))
width = 0.3

plt.bar(x - width/2, group1, width, label='Group 1', color='blue')
plt.bar(x + width/2, group2, width, label='Group 2', color='orange')

plt.xticks(x, categories)
plt.title("Grouped Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()
plt.show()
```

####  Basic Horizontal Bar Plot
* A horizontal variant of the bar plot, better for readability when category names are long.
* plt.barh() creates horizontal bars.
* Values are now on the x-axis, and categories are on the y-axis.


```python
plt.barh(categories, values, color='lightgreen')
plt.title("Horizontal Bar Plot")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

#### Stacked Bar Plot
* Used to visualize parts of a whole within a category.
* The bottom parameter stacks one set of bars on top of another.
* Shows the cumulative value for each category.

```python
values1 = [10, 15, 7, 12]
values2 = [8, 14, 6, 10]

plt.bar(categories, values1, label='Group 1', color='blue')
plt.bar(categories, values2, bottom=values1, label='Group 2', color='orange')

plt.title("Stacked Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()
plt.show()
```

#### Histogram Chart

* A histogram in Matplotlib is a graphical representation of the distribution of numerical data, used to visualize the frequency of data points within specified intervals (bins).

```python
# Generate random data
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, StdDev=10, 1000 points
data
```

#### Basic plot
* A standard histogram to visualize the frequency distribution of data.
* When to use: To observe the overall shape and spread of your dataset.


```python
plt.hist(data,bins=5)
plt.show()
```

* Histogram with Custom Number of Bins
* When to use: When you want more or fewer intervals (bins) to better analyze specific data patterns.

* Normalized Histogram (Probability Density)
* Displays the data as probabilities rather than raw counts, making it easier to compare datasets.
* When to use: When you want to compare datasets of different sizes.

```python
plt.hist(data, density=True)
plt.show()
```

#### Stacked Histogram
* Displays multiple datasets stacked on top of each other.
* When to use: To compare the cumulative contribution of different datasets.

```python
data1 = np.random.normal(loc=50, scale=10, size=1000)
data2 = np.random.normal(loc=60, scale=15, size=1000)
```

```python
plt.hist([data1, data2], bins=10, stacked=True)
plt.show()
```

#### Histogram with Custom Colors
* Customizes the histogram's appearance using color options.
* When to use: When you want to enhance the visualization for presentations or clarity.

```python
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.show()
```

#### Horizontal Histogram
* Displays the histogram horizontally instead of vertically.
* When to use: When horizontal orientation better suits the context or readability of your chart.

```python
plt.hist(data, bins=10, orientation='horizontal')
plt.show()
```

#### Histogram with Log Scale
* Plots frequencies on a logarithmic scale, useful for data with large ranges.
* When to use: For datasets with exponential distribution or wide data ranges.

```python
plt.hist(data, bins=10, log=True)
plt.show()
```

#### Overlaid Histograms
* Plots multiple datasets with some transparency to observe overlaps.
* When to use: To compare distributions of different datasets side-by-side.

```python
plt.hist(data1, bins=10, alpha=0.5, label='Data 1')
plt.hist(data2, bins=10, alpha=0.5, label='Data 2')
plt.legend()
plt.show()
```

#### Cumulative Histogram
* Shows cumulative frequencies instead of regular frequencies.
* When to use: To understand how data accumulates over intervals.

```python
plt.hist(data, bins=10, cumulative=True)
plt.show()
```

#### Step Histogram (Unfilled)
* Creates an outline of the histogram bins without filling them.
* When to use: When you want a cleaner look or to avoid overlaps in multiple histograms.

```python
plt.hist(data, bins=10, histtype='step', color='red')
plt.show()
```

#### Step-filled Histogram
* Similar to the step histogram, but with filled bins for better visibility.
* When to use: When you want the benefits of a step histogram with color filling.

```python
plt.hist(data, bins=10, histtype='stepfilled', alpha=0.5)
plt.show()
```

#### Custom Bin Ranges
* Define custom intervals (bins) for the histogram.
* When to use: To focus on specific ranges of data or when the intervals have special significance.

```python
bin_edges = [0, 10, 20, 30, 40, 50]
plt.hist(data, bins=bin_edges)
plt.show()
```

#### Pie Chart

#### Basics pie chart
* Purpose: A straightforward chart showing how different parts contribute to a whole.
* Key Feature: Each slice represents a portion of the total.

```python
labels_name = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels_name)
plt.title("Basic Pie Chart")
plt.show()
```

#### Pie Chart with Custom Colors
* Purpose: Makes the chart visually appealing by specifying custom colors.
* Key Feature: Assigns colors to slices.

```python
colors = ['gold', 'lightcoral', 'lightskyblue', 'yellowgreen']
plt.pie(sizes, labels=labels_name, colors=colors, autopct='%1.1f%%')
plt.title("Pie Chart with Custom Colors")
plt.show()
```

#### Exploded Pie Chart
* Purpose: To emphasize a specific slice by pulling it out slightly.
* Key Feature: Makes the highlighted slice stand out visually.

```python
explode = (0, 0.1, 0, 0)  # Explodes the second slice (B)
plt.pie(sizes, explode=explode, labels=labels_name, autopct='%1.1f%%')
plt.title("Exploded Pie Chart")
plt.show()
```

#### Pie Chart with Percentages
* Purpose: Shows the contribution of each slice as a percentage.
* Key Feature: Adds percentage labels to the chart for better clarity.
* The autopct='%1.1f%%' formats the percentage values with 1 decimal place.

```python
plt.pie(sizes, labels=labels_name, autopct='%1.1f%%',radius=.5)
plt.title("Pie Chart with Percentages")
plt.show()
```

#### Start Angle Adjusted Pie Chart
* Purpose: Rotates the starting point of the chart.
* Key Feature: Adjusts startangle to set which slice starts at the top.
* Explanation: startangle=90 rotates the chart so the first slice starts at 12 o’clock.

```python
plt.pie(sizes, labels=labels_name, startangle=90, autopct='%1.1f%%',counterclock=True,rotatelabels=True)
plt.title("Pie Chart with Start Angle Adjusted")
plt.legend()
plt.show()
```

#### Doughnut Chart (Ring Chart)
* Purpose: A pie chart with a hole in the middle to create a "doughnut" shape.
* Key Feature: Adds a modern and clean look.

```python
plt.pie(sizes, labels=labels_name, autopct='%1.1f%%', wedgeprops={'width': 0.6})
plt.title("Doughnut Chart")
plt.show()
```

#### Nested Pie Chart
* Purpose: Displays hierarchical data or two levels of information.
* Key Feature: Multiple rings represent different layers of data.


```python
inner_sizes = [7, 8, 9, 6]
outer_sizes = [10, 15, 20, 25]

fig, ax = plt.subplots()
ax.pie(outer_sizes, radius=1, labels=labels_name, autopct='%1.1f%%')
ax.pie(inner_sizes, radius=0.7, labels=labels_name, autopct='%1.1f%%')
plt.title("Nested Pie Chart")
plt.show()
```

#### Shadowed Pie Chart
* Purpose: Adds a shadow to the chart for a 3D-like effect.
* Key Feature: Adds depth to the chart.
* Use shadow=True to create the effect.

```python
plt.pie(sizes, labels=labels_name, shadow=True, autopct='%1.1f%%')
plt.title("Shadowed Pie Chart")
plt.show()
```

#### Custom Wedge Styles
* Purpose: Modify the edges and style of wedges for a unique look.
* Key Feature: Customize wedge borders, colors, or patterns.
* Use wedgeprops to define edge properties like color and width.

```python
plt.pie(
    sizes,
    labels=labels_name,
    autopct='%1.1f%%',
    wedgeprops={'edgecolor': 'black', 'linewidth': 2}
)
plt.title("Custom Wedge Styled Pie Chart")
plt.show()
```

#### Sub Plots
The subplot() function in Matplotlib allows you to create a grid of plots within a single figure.

*  Basic syntax
*  matplotlib.pyplot.subplot(nrows, ncols, index, **kwargs)
*  nrows --> The number of rows in the subplot grid. Example: subplot(2, 3, 1) means 2 rows in the grid.
*  ncols --> The number of columns in the subplot grid. Example: subplot(2, 3, 1) means 3 columns in the grid.
*  index -- > The position of the current subplot in the grid, indexed from 1
*  example --> subplot(2, 3, 1) places the plot in the top-left of a 2x3 grid.
*  subplot(2, 3, 6) places it in the bottom-right of the grid.
*  fig, axs = plt.subplots(2, 1, sharex=True)
*  Sharing Axes, the sharex=True parameter is used when creating multiple subplots to share the same x-axis. This means the x-axis values will be consistent across the subplots, making it easier to compare data visually. Similarly, you can use sharey=True to share the y-axis across the subplots.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)  # Top-left
plt.plot(x, np.sin(x))
plt.title("Sin(x)")

plt.subplot(2, 2, 2)  # Top-right
plt.plot(x, np.cos(x))
plt.title("Cos(x)")

plt.subplot(2, 2, 3)  # Bottom-left
plt.plot(x, np.tan(x))
plt.title("Tan(x)")

plt.subplot(2, 2, 4)  # Bottom-right
plt.plot(x, np.exp(-x))
plt.title("Exp(-x)")

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, sharex=True)

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

axs[0].plot(x, y1, label='sin(x)')
axs[0].legend()

axs[1].plot(x, y2, label='cos(x)')
axs[1].legend()

# Shared x-axis
plt.xlabel("X-Axis")
plt.show()
```

##### Polar 
* When using the polar=True parameter in Matplotlib, the subplot switches from a Cartesian coordinate system (x and y axes) to a polar coordinate system.
* Data is represented in terms of angles ( 𝜃 θ) and radii ( 𝑟 r).
* Polar plots are commonly used for circular data, such as periodic functions, compass directions, or radar charts.

```python
# Cartesian coordinates
x = np.array([1, 2, 3, 4])
y = np.array([5, 0, 3, 4])

# Convert to polar coordinates
r = np.sqrt(x**2 + y**2)              # Calculate radius
theta = np.arctan2(y, x)              # Calculate angle (in radians)

# Create polar plot
plt.subplot(1 , 1, 1, polar=True)
plt.plot(theta, r, marker='o', label="Polar Points")
plt.title("Polar Plot from Cartesian Coordinates", va='bottom')
plt.legend()
plt.show()
```

* If you have values in Cartesian coordinates, like 𝑥 = [ 1 , 2 , 3 , 4 ] x=[1,2,3,4] and 𝑦 = [ 5 , 0 , 3 , 4 ] y=[5,0,3,4], and you want to create a polar plot, you need to convert Cartesian coordinates ( 𝑥 , 𝑦 x,y) into polar coordinates ( 𝑟 , 𝜃 r,θ).

```python
# Define a figure with 2x2 grid
plt.subplot(2, 2, 1)  # First subplot
plt.title("Subplot 1")
plt.plot([1, 2, 3], [1, 4, 9])

plt.subplot(2, 2, 2)  # Second subplot
plt.title("Subplot 2")
plt.plot([1, 2, 3], [9, 4, 1])

plt.subplot(2, 2, 3)  # Third subplot
plt.title("Subplot 3")
plt.plot([1, 2, 3], [3, 3, 3])

plt.subplot(2, 2, 4, polar=True)  # Fourth subplot (polar)
plt.title("Polar Subplot")
plt.plot([0, 1, 2], [0, 1, 2])

plt.tight_layout()
plt.show()
```
